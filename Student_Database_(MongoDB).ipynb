# **1) Find the student name who scored maximum scores in all (exam, quiz and homework)?**


```python
list_all_marks=[]
for i in x.find():
  total=0
  for j in range(len(i['scores'])):
    total=total+i['scores'][j]['score']
  list_all_marks.append(total)  
  x.update_one({'_id': i['_id']},{'$set': {'total': total}})
```

```python
for i in x.find():
  if(i['total']==max(list_all_marks)):
    print(i)
    break
```

# 3.Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

```python
# less than 40 in any one of the score then consider as fail
# greater than 40 in  all of the score then consider as pass
for i in x.find():
  if (i['scores'][0]['score']>=40 and i['scores'][1]['score']>=40 and i['scores'][2]['score']>=40):
    x.update_one({'_id': i['_id']},{'$set': {'grade': 'pass'}})
  else:
    x.update_one({'_id': i['_id']},{'$set': {'grade': 'fail'}})
```

```python
for i in x.find():
  print(i)
```

# 4) Find the total and average of the exam, quiz and homework and store them in a separate collection.


```python
y = db['qs_4']
```

```python
list_exam_mark=[]
list_quiz_marks=[]
list_homework_marks=[]
for i in x.find():
  list_exam_mark.append(i['scores'][0]['score'])
  list_quiz_marks.append(i['scores'][1]['score'])
  list_homework_marks.append(i['scores'][2]['score'])

avg_exam=sum(list_exam_mark)/len(list_exam_mark)
avg_quiz=sum(list_quiz_marks)/len(list_quiz_marks)
avg_homework=sum(list_homework_marks)/len(list_homework_marks)
```

```python
data_1={
    'all_marks' :[{'sum_exam_mk': sum(list_exam_mark)},{'sum_quiz_mk' :sum(list_quiz_marks)},{'sum_hw_mk' :sum(list_homework_marks)}],
    'all_marks_avg' :[{'avg_exam' :avg_exam},{'avg_quiz' :avg_quiz},{'avg_hw' :avg_homework}]
       }
```

```python
data_1
```

```python
y.insert_one(data_1)
```
